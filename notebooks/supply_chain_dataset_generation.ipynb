## Retail Supply Chain Dataset

This dataset simulates a retail distribution network including warehouse operations, product demand, and transportation costs.

### Key Variables

- product category
- units sold
- inventory levels
- warehouse utilization
- transportation cost per unit
- regional revenue

### Analytics Use Case

The dataset supports analysis of:

- revenue trends over time
- backorders by product category
- warehouse capacity utilization
- regional demand patterns

These metrics reflect typical analytics used by retail operations and supply chain managers.

```python
import pandas as pd
import numpy as np
from datetime import datetime

# ------------------------------
# STEP 1: Set random seed
# ------------------------------
np.random.seed(42)

# ------------------------------
# STEP 2: Define lookup lists
# ------------------------------
dates = pd.date_range(start="2023-01-01", end="2025-12-31", freq="D")

regions = ["Northeast", "South", "Midwest", "West"]
warehouses = ["DC_Dallas", "DC_Atlanta", "DC_Chicago", "DC_Phoenix", "DC_NewJersey"]
categories = ["Apparel", "Electronics", "Home Goods", "Beauty", "Sporting Goods"]
suppliers = ["Supplier_A", "Supplier_B", "Supplier_C", "Supplier_D", "Supplier_E"]

products = {
    "Apparel": ["Shirts", "Jeans", "Jackets", "Socks"],
    "Electronics": ["Headphones", "Tablets", "Speakers", "Chargers"],
    "Home Goods": ["Storage Bins", "Lamp Sets", "Cookware", "Bedding"],
    "Beauty": ["Skin Care", "Makeup Kits", "Hair Products", "Fragrances"],
    "Sporting Goods": ["Yoga Mats", "Dumbbells", "Resistance Bands", "Water Bottles"]
}

# ------------------------------
# STEP 3: Set number of rows
# ------------------------------
num_rows = 3000

# ------------------------------
# STEP 4: Generate synthetic data
# ------------------------------
data = []

for _ in range(num_rows):
    date = np.random.choice(dates)
    region = np.random.choice(regions)
    warehouse = np.random.choice(warehouses)
    category = np.random.choice(categories)
    product = np.random.choice(products[category])
    supplier = np.random.choice(suppliers)

    units_sold = np.random.randint(20, 500)
    inventory_level = np.random.randint(100, 5000)
    reorder_point = np.random.randint(150, 1200)
    lead_time_days = np.random.randint(2, 21)

    transportation_cost = round(np.random.uniform(150.0, 5000.0), 2)
    unit_price = round(np.random.uniform(8.0, 250.0), 2)

    warehouse_capacity = np.random.randint(8000, 15000)
    warehouse_used = np.random.randint(4000, warehouse_capacity + 1)

    # Order status logic
    status_roll = np.random.rand()
    if inventory_level < reorder_point and status_roll < 0.45:
        order_status = "Backordered"
        backorder_units = np.random.randint(5, 150)
    elif status_roll < 0.88:
        order_status = "Fulfilled"
        backorder_units = 0
    elif status_roll < 0.95:
        order_status = "Pending"
        backorder_units = 0
    else:
        order_status = "Canceled"
        backorder_units = 0

    data.append([
        date,
        region,
        warehouse,
        category,
        product,
        supplier,
        units_sold,
        inventory_level,
        reorder_point,
        lead_time_days,
        transportation_cost,
        order_status,
        backorder_units,
        warehouse_capacity,
        warehouse_used,
        unit_price
    ])

# ------------------------------
# STEP 5: Create DataFrame
# ------------------------------
columns = [
    "Date",
    "Region",
    "Warehouse",
    "Category",
    "Product",
    "Supplier",
    "Units_Sold",
    "Inventory_Level",
    "Reorder_Point",
    "Lead_Time_Days",
    "Transportation_Cost",
    "Order_Status",
    "Backorder_Units",
    "Warehouse_Capacity",
    "Warehouse_Used",
    "Unit_Price"
]

df = pd.DataFrame(data, columns=columns)

# ------------------------------
# STEP 6: Add calculated columns
# ------------------------------
df["Sales_Revenue"] = (df["Units_Sold"] * df["Unit_Price"]).round(2)
df["Inventory_Position"] = df["Inventory_Level"] - df["Reorder_Point"]
df["Warehouse_Utilization_Pct"] = ((df["Warehouse_Used"] / df["Warehouse_Capacity"]) * 100).round(2)
df["Transport_Cost_Per_Unit"] = (df["Transportation_Cost"] / df["Units_Sold"]).round(2)

# ------------------------------
# STEP 7: Save to CSV
# ------------------------------
file_name = "retail_distribution_dataset.csv"
df.to_csv(file_name, index=False)

# ------------------------------
# STEP 8: Preview data
# ------------------------------
print("Dataset created successfully.")
print("File saved as:", file_name)
print("\nShape of dataset:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
```

```python
# ------------------------------
# STEP 2: Load dataset
# ------------------------------

import pandas as pd

df = pd.read_csv("retail_distribution_dataset.csv")

print("Dataset loaded.")
print(df.shape)


# ------------------------------
# KPI 1: Total Revenue
# ------------------------------

total_revenue = df["Sales_Revenue"].sum()


# ------------------------------
# KPI 2: Total Units Sold
# ------------------------------

total_units = df["Units_Sold"].sum()


# ------------------------------
# KPI 3: Fulfillment Rate
# ------------------------------

fulfilled_orders = len(df[df["Order_Status"] == "Fulfilled"])
total_orders = len(df)

fulfillment_rate = fulfilled_orders / total_orders


# ------------------------------
# KPI 4: Backorder Rate
# ------------------------------

backorders = len(df[df["Order_Status"] == "Backordered"])

backorder_rate = backorders / total_orders


# ------------------------------
# KPI 5: Average Lead Time
# ------------------------------

avg_lead_time = df["Lead_Time_Days"].mean()


# ------------------------------
# KPI 6: Average Warehouse Utilization
# ------------------------------

avg_utilization = df["Warehouse_Utilization_Pct"].mean()


# ------------------------------
# KPI 7: Transportation Cost per Unit
# ------------------------------

avg_transport_cost_unit = df["Transport_Cost_Per_Unit"].mean()


# ------------------------------
# KPI 8: Inventory Risk
# ------------------------------

inventory_risk = len(df[df["Inventory_Position"] < 0])

inventory_risk_rate = inventory_risk / total_orders


# ------------------------------
# Display KPI Results
# ------------------------------

print("\n--- SUPPLY CHAIN KPI SUMMARY ---")

print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Total Units Sold: {total_units:,}")

print(f"\nFulfillment Rate: {fulfillment_rate:.2%}")
print(f"Backorder Rate: {backorder_rate:.2%}")

print(f"\nAverage Lead Time: {avg_lead_time:.2f} days")
print(f"Average Warehouse Utilization: {avg_utilization:.2f}%")

print(f"\nAverage Transportation Cost Per Unit: ${avg_transport_cost_unit:.2f}")

print(f"\nInventory Risk Rate: {inventory_risk_rate:.2%}")
```

```python
# ------------------------------
# STEP 3: Create Aggregated Data
# ------------------------------

# Monthly sales trend
monthly_sales = df.groupby(pd.to_datetime(df["Date"]).dt.to_period("M"))["Sales_Revenue"].sum().reset_index()
monthly_sales["Date"] = monthly_sales["Date"].astype(str)

# Sales by region
sales_by_region = df.groupby("Region")["Sales_Revenue"].sum().reset_index()

# Sales by category
sales_by_category = df.groupby("Category")["Sales_Revenue"].sum().reset_index()

# Warehouse utilization
warehouse_utilization = df.groupby("Warehouse")["Warehouse_Utilization_Pct"].mean().reset_index()

# Transportation cost by region
transport_cost_region = df.groupby("Region")["Transport_Cost_Per_Unit"].mean().reset_index()

# Backorders by category
backorders_category = df[df["Order_Status"] == "Backordered"].groupby("Category")["Backorder_Units"].sum().reset_index()


# ------------------------------
# Export Aggregated Files
# ------------------------------

monthly_sales.to_csv("monthly_sales.csv", index=False)
sales_by_region.to_csv("sales_by_region.csv", index=False)
sales_by_category.to_csv("sales_by_category.csv", index=False)
warehouse_utilization.to_csv("warehouse_utilization.csv", index=False)
transport_cost_region.to_csv("transport_cost_region.csv", index=False)
backorders_category.to_csv("backorders_category.csv", index=False)

print("Aggregated datasets created for Tableau dashboard.")
```

## Military Logistics Dataset

This dataset simulates global logistics operations for military supply movement.

### Key Variables

- Shipment date
- Logistics hub location
- Operational theater
- Carrier mode (Air, Ground, Sea)
- Units shipped
- Transit time
- Equipment readiness rate

### Analytics Use Case

This dataset supports dashboards analyzing:

- logistics hub throughput
- shipment distribution by theater
- transportation mode utilization
- operational readiness metrics

The data simulates realistic logistics flows between major military supply hubs.

```python
import pandas as pd
import numpy as np

np.random.seed(42)

# date range
dates = pd.date_range(start="2023-01-01", end="2025-12-31", freq="D")

# military logistics fields
theaters = ["CENTCOM", "EUCOM", "INDOPACOM", "AFRICOM"]
hubs = ["Kuwait Hub", "Ramstein Hub", "Guam Hub", "Djibouti Hub"]

shipment_types = ["Equipment", "Medical", "Fuel", "Food", "Ammunition"]

equipment_categories = [
    "Vehicles",
    "Communications",
    "Medical Supplies",
    "Engineering Equipment",
    "Weapons Systems"
]

carrier_modes = ["Air", "Sea", "Ground"]

mission_priority = ["High", "Medium", "Low"]

readiness_status = ["Ready", "Maintenance", "Deployed"]

rows = 2500

data = []

for i in range(rows):

    date = np.random.choice(dates)
    theater = np.random.choice(theaters)
    hub = np.random.choice(hubs)

    container = f"CNT-{np.random.randint(100000,999999)}"

    shipment = np.random.choice(shipment_types)
    equipment = np.random.choice(equipment_categories)

    units = np.random.randint(10,200)

    transit_days = np.random.randint(2,25)

    delay_flag = np.random.choice([0,1], p=[0.85,0.15])

    readiness = np.random.choice(readiness_status)

    inventory = np.random.randint(50,2000)

    priority = np.random.choice(mission_priority)

    carrier = np.random.choice(carrier_modes)

    data.append([
        date,
        theater,
        hub,
        container,
        shipment,
        equipment,
        units,
        transit_days,
        delay_flag,
        readiness,
        inventory,
        priority,
        carrier
    ])

columns = [
    "Date",
    "Theater",
    "Logistics_Hub",
    "Container_ID",
    "Shipment_Type",
    "Equipment_Category",
    "Units_Shipped",
    "Transit_Days",
    "Delay_Flag",
    "Readiness_Status",
    "Inventory_On_Hand",
    "Mission_Priority",
    "Carrier_Mode"
]

df_military = pd.DataFrame(data, columns=columns)

df_military.to_csv("military_logistics_dataset.csv", index=False)

print("Dataset Created")
print(df_military.shape)
df_military.head()
```

```python
# ------------------------------
# Military Logistics KPI Analysis
# ------------------------------

import pandas as pd

df = pd.read_csv("military_logistics_dataset.csv")

# KPI 1 - Total Units Shipped
total_units = df["Units_Shipped"].sum()

# KPI 2 - Average Transit Time
avg_transit = df["Transit_Days"].mean()

# KPI 3 - Delay Rate
delay_rate = df["Delay_Flag"].mean()

# KPI 4 - Readiness Rate
ready_count = len(df[df["Readiness_Status"] == "Ready"])
readiness_rate = ready_count / len(df)

# KPI 5 - Inventory by Theater
inventory_by_theater = df.groupby("Theater")["Inventory_On_Hand"].sum()

print("MILITARY LOGISTICS KPI SUMMARY\n")

print(f"Total Units Shipped: {total_units:,.0f}")
print(f"Average Transit Time: {avg_transit:.2f} days")
print(f"Shipment Delay Rate: {delay_rate:.2%}")
print(f"Equipment Readiness Rate: {readiness_rate:.2%}")

print("\nInventory by Theater")
print(inventory_by_theater)
```

```python
# ------------------------------
# Create Tableau-ready datasets
# ------------------------------

# Units shipped by theater
units_by_theater = df.groupby("Theater")["Units_Shipped"].sum().reset_index()

# Units shipped by logistics hub
units_by_hub = df.groupby("Logistics_Hub")["Units_Shipped"].sum().reset_index()

# Transit time trend
transit_trend = df.groupby("Date")["Transit_Days"].mean().reset_index()

# Carrier usage
carrier_usage = df.groupby("Carrier_Mode")["Units_Shipped"].sum().reset_index()

# Readiness distribution
readiness_dist = df.groupby("Readiness_Status")["Units_Shipped"].sum().reset_index()

# Save files
units_by_theater.to_csv("units_by_theater.csv", index=False)
units_by_hub.to_csv("units_by_hub.csv", index=False)
transit_trend.to_csv("transit_trend.csv", index=False)
carrier_usage.to_csv("carrier_usage.csv", index=False)
readiness_dist.to_csv("readiness_distribution.csv", index=False)

print("Aggregated datasets for Tableau created successfully.")
```

## Healthcare Supply Chain Dataset

This dataset simulates medical supply demand across hospital networks in the United States.

### Key Variables

- hospital location
- medical supply category
- units used
- supplier lead time
- inventory levels
- total supply spend

### Analytics Use Case

The dataset supports dashboards analyzing:

- hospital supply demand by location
- supplier delivery performance
- medical supply utilization patterns
- healthcare supply spend by category

This type of analysis is commonly used in hospital procurement and healthcare supply chain operations.

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

np.random.seed(42)

# hospitals with map coordinates
hospitals = [
("Mayo Clinic","Rochester","MN",44.0225,-92.4699),
("Cleveland Clinic","Cleveland","OH",41.5036,-81.6206),
("UT Southwestern","Dallas","TX",32.8136,-96.8404),
("Mass General","Boston","MA",42.3626,-71.0695),
("Johns Hopkins","Baltimore","MD",39.2975,-76.5929),
("UCLA Medical Center","Los Angeles","CA",34.0655,-118.4468),
("Houston Methodist","Houston","TX",29.7093,-95.3982),
("Northwestern Memorial","Chicago","IL",41.8947,-87.6216),
("Emory University Hospital","Atlanta","GA",33.7925,-84.3200),
("Stanford Health Care","Palo Alto","CA",37.4336,-122.1750)
]

# supply categories
basic_supplies = ["Gloves","Surgical Masks","IV Bags","Syringes","Saline"]
clinical_supplies = ["Catheters","Surgical Kits","Blood Tubes","Dressings"]
critical_equipment = ["Ventilators","Infusion Pumps","Defibrillators","Pacemakers"]

suppliers = ["Medline","Cardinal Health","McKesson","BD","3M"]

rows = []

start_date = datetime(2024,1,1)

for i in range(2500):

    hospital = random.choice(hospitals)
    
    category_choice = random.choice(["Basic","Clinical","Critical"])
    
    if category_choice == "Basic":
        item = random.choice(basic_supplies)
        units = np.random.randint(1000,5000)
        cost = round(np.random.uniform(0.2,5),2)
        
    elif category_choice == "Clinical":
        item = random.choice(clinical_supplies)
        units = np.random.randint(200,1200)
        cost = round(np.random.uniform(10,300),2)
        
    else:
        item = random.choice(critical_equipment)
        units = np.random.randint(10,80)
        cost = round(np.random.uniform(2000,15000),2)

    inventory = np.random.randint(100,10000)
    reorder = np.random.randint(200,2000)
    lead = np.random.randint(2,30)

    rows.append([
        start_date + timedelta(days=random.randint(0,365)),
        hospital[0],
        hospital[1],
        hospital[2],
        hospital[3],
        hospital[4],
        category_choice,
        item,
        units,
        cost,
        inventory,
        reorder,
        random.choice(suppliers),
        lead
    ])

columns = [
"Date","Hospital","City","State","Latitude","Longitude",
"Supply_Category","Supply_Item","Units_Used","Unit_Cost",
"Inventory_On_Hand","Reorder_Point","Supplier","Lead_Time_Days"
]

df_health = pd.DataFrame(rows,columns=columns)

df_health["Total_Spend"] = df_health["Units_Used"] * df_health["Unit_Cost"]

df_health.to_csv("healthcare_supply_chain_dataset.csv",index=False)

print("Healthcare dataset created:",df_health.shape)
df_health.head()
```

```python
# ------------------------------
# Create Tableau-ready datasets
# ------------------------------

# Total spend by supply category
spend_by_category = df_health.groupby("Supply_Category")["Total_Spend"].sum().reset_index()

# Units used by supply item
usage_by_item = df_health.groupby("Supply_Item")["Units_Used"].sum().reset_index()

# Supplier lead time analysis
supplier_lead = df_health.groupby("Supplier")["Lead_Time_Days"].mean().reset_index()

# Hospital inventory risk
inventory_risk = df_health.groupby("Hospital")["Inventory_On_Hand"].mean().reset_index()

# Supply usage by hospital for map
hospital_usage = df_health.groupby(
    ["Hospital","City","State","Latitude","Longitude"]
)["Units_Used"].sum().reset_index()

# Save files
spend_by_category.to_csv("health_spend_category.csv",index=False)
usage_by_item.to_csv("health_usage_item.csv",index=False)
supplier_lead.to_csv("health_supplier_leadtime.csv",index=False)
inventory_risk.to_csv("health_inventory_risk.csv",index=False)
hospital_usage.to_csv("health_hospital_usage.csv",index=False)

print("Healthcare Tableau datasets created successfully.")
```
